```python
import matplotlib.pyplot as plt
from six.moves import cPickle as pickle
import numpy as np

with open('rankings.pickle', 'rb') as pf:
    rankings = pickle.load(pf)
with open('counts.pickle', 'rb') as pf:
    sorted_words = pickle.load(pf)
```

```python
# # Option A: freq. in x vs freq. in y
x = []
y = []
max_count = 1000
# to_look = sorted_words[0:max_count - 1]
for entry in sorted_words:
    word = entry[0]
    ranking = rankings[word]
    x.append(ranking[0])
    y.append(ranking[1])
x = np.array(x)
y = np.array(y)
x = np.power(x, .99)
y = np.power(y, .99)
x /= max(x)
y /= max(y)
plt.plot(x, y, 'o')
# plt.plot([0,1],[0,1], 'k-')
plt.axis('equal')
# plt.axis([0, 1, 0, 1])
plt.title('Word Frequencies in Scientific and Newspaper Corpuses')
plt.xlabel('Frequency in Newspapers (normalized)')
plt.ylabel('Frequency in Scientific Journals (normalized)')
plt.savefig('versionA.png')
plt.show()
```
